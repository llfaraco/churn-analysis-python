# üìâ Churn Analysis ‚Äì EDA em PythonAn√°lise explorat√≥ria de churn com foco em comportamento, risco e padr√µes de cancelamento.

```python
# Imports principais e leitura do datasetimport pandas as pdimport numpy as npimport matplotlib.pyplot as pltimport seaborn as snssns.set(style="whitegrid")plt.rcParams['figure.figsize'] = (10, 6)df = pd.read_csv("../data/churn_clientes.csv")df.head()
```

```python
# Estrutura geral do datasetdf.info()
```

```python
# Estat√≠sticas descritivasdf.describe(include="all")
```

```python
# C√°lculo da taxa geral de churnchurn_rate = df['churn'].mean()print(f"Taxa geral de churn: {churn_rate:.2%}")
```

```python
# Taxa de churn por canalcanal_churn = df.groupby('canal')['churn'].mean().sort_values(ascending=False)canal_churn
```

```python
# Visualiza√ß√£o da taxa de churn por canalsns.barplot(x=canal_churn.index, y=canal_churn.values)plt.title('Taxa de churn por canal')plt.ylabel('Taxa de churn')plt.xlabel('Canal')plt.xticks(rotation=0)plt.tight_layout()plt.show()
```

```python
# Rela√ß√£o entre atrasos de pagamento e churnsns.boxplot(x='churn', y='atrasos_pagamento', data=df)plt.title('Atrasos de pagamento por status de churn')plt.xlabel('Churn (0 = ativo, 1 = cancelado)')plt.ylabel('Atrasos de pagamento')plt.tight_layout()plt.show()
```

```python
# Matriz de correla√ß√£o entre vari√°veis num√©ricascorr = df[['idade','tempo_de_casa_meses','ticket_medio','atrasos_pagamento','atendimentos_suporte','churn']].corr()sns.heatmap(corr, annot=True, cmap='Blues')plt.title('Matriz de correla√ß√£o')plt.tight_layout()plt.show()
```

## ‚úÖ Conclus√µes- Clientes com maior n√∫mero de atrasos de pagamento apresentam maior probabilidade de churn.- Segmentos com menor ticket m√©dio concentram proporcionalmente mais cancelamentos.- Queda de uso do produto est√° associada a maior risco de cancelamento.- Clientes com muitos atendimentos de suporte tendem a cancelar mais.Este tipo de an√°lise pode apoiar times de Customer Success, Marketing e Produto a reduzir churn e aumentar a previsibilidade de receita.
